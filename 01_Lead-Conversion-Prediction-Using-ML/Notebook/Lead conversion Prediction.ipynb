```
import pandas as pd
df = pd.read_csv("lead_conversion_dataset.csv")
print(df)
```

```
X = df[["Age","WebsiteVisits","EmailOpens"]]
y = df["Converted"]
print(X.head())
print(y.head())
```

```
#From sklearn library
#take model_selection module
#import train_test_split function

#X = Input Features
#y = Output/Answer

#X_train = Questions for study
#y_train = Answers for study
#X_test = Questions in exam
#y_test = Real answers of exam

#20% Testing
#80% Training

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
)
print("X_train")
print(X_train)

print("\nX_test")
print(X_test)

print("\ny_train")
print(y_train)

print("\ny_test")
print(y_test)
```

```
#From sklearn   -  library
#Import LogisticRegression   - algorithm

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```
# Teach the model for prediction

model.fit(X_train, y_train)
```

```
# model.predict(X_test)  - Ask the model to answer the test questions.

y_pred = model.predict(X_test)
print("Actual Values:")
print(y_test.values)

print("Predicted Values:")
print(y_pred)
```

```
# check the accuracy that is how many answer were correct

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy =", accuracy)
```

```
# Check the New Lead whether the model works correctly or Not

rajesh = [[35, 2, 10 ]]

result = model.predict(rajesh)

print(result)
```
